```python
#مقادیر ستون نمره
df['نمره'].unique()
```

```python
#پاکسازی دیتای نمره
def Score_conversion(x):
    try:
        return float(x)
    except:
        return None

df['نمره_عددی'] = df['نمره'].apply(Score_conversion)
df['نمره_عددی'].describe()
```

```python
plt.Figure(figsize=(10, 5))
sns.histplot(df['نمره_عددی'].dropna() , bins=20)
plt.title('توزیع نمرات دانشجویان')
plt.xlabel('نمره')
plt.ylabel('تعداد')
plt.show()
```

```python
zero_scores = df[df['نمره_عددی'] == 0]
zero_scores['وضعيت‌ترم‌دانشجو'].value_counts()
```

```python
df['نمره_مشکوک'] = (
    (df['نمره_عددی'] == 0) &
    (df['وضعيت‌ترم‌دانشجو'] == 'عادي')
).astype(int)

df['نمره_مشکوک'].sum()
```
